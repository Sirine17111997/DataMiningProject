```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
print(df.shape)
df.head(50)
```

```python
df.info()
```

```python
#reduire le nombre des parmetres 
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
print(df.shape)
df.head()
```

```python
#extraire le variable à predire
df1= df.drop('Survived',axis='columns')
survie=df.Survived
survie
```

```python
#changer le contenu de la varible sex par(male=1,female=2)
df1.Sex = df1.Sex.map({'male': 1, 'female': 2})
```

```python
df1.head()
```

```python
df1.Age[:20]
```

```python
#on change les NaN par la valeur moyenne 
df1.Age = df1.Age.fillna(df1.Age.mean())
df1.Age[:20]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Partition en échantillons d’apprentissage et de test
X_train, X_test, y_train, y_test = train_test_split(df1,survie,test_size=0.2)
```

```python
#importer la librairie tree
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#affichage graphique de l'arbre 
tree.plot_tree(model)
```
